```python
#Multiple regression helps against outliers. 

#Prediction of Carbon Monoxide emission of a car
import pandas as pd
from sklearn import linear_model

#import the data set with the pandas method
df= pd.read_csv('data.csv',header=0, delim_whitespace=True)
#headers= ['Car', 'Model', 'Volume', 'Weight', 'CO2']
#df.columns= headers
#df.to_csv('newsource.csv')

#List the independent and variable
x= df [['Weight',  'Volume']]
y= df['CO2']

#Introducing the fit() method that takes dependent and independent values as parameters and fill the regression object
regr= linear_model.LinearRegression()
regr.fit(x,y)

#Predict the CO**2 emission of a car where the weight is 2300kg and the volume is 1300cm**3
predictedCO2= regr.predict([[2300, 1300]])

print(predictedCO2)
```

```python
#predict the CO2 emission from a 1.3ltre car that weighs 2300kg. scaled data that was regressed based on coefficient

import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale= StandardScaler()

df= pd.read_csv('data.csv')

x= df[['Weight', 'Volume']]
y= df[['CO2']]

scaledX= scale.fit_transform(x)

regr= linear_model.LinearRegression()
regr.fit(scaledX, y)

scaled= scale.transform([[2300, 1.3]])

predictedCO2= regr.predict ([scaled[0]])

print(predictedCO2)


```
